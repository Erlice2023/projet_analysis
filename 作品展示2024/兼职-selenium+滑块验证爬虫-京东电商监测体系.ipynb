```python
import difflib
# import jieba
import os
import Levenshtein
import numpy as np
from selenium.webdriver.common.action_chains import ActionChains
import requests
import re
import io
import urllib.request
import pandas as pd
import requests
import pytesseract
from PIL import Image
from selenium import webdriver # 导入webdriver包
from selenium.webdriver import ActionChains
import time 
```

```python
chance=0
# global chance
# global driver
def login(canshu):     #店铺登陆函数
    global chance
    global driver
    try:
        driver = webdriver.Chrome(r"Chrome\chromedriver.exe")
        driver.get("https://xxx.action")
        driver.find_element_by_id("account-login").click()
        time.sleep(5)
        driver.switch_to_frame("loginFrame")
        driver.find_element_by_id("loginname").send_keys("xxx")
        driver.find_element_by_id('nloginpwd').send_keys("xxx")
        time.sleep(5)
        driver.find_element_by_id("paipaiLoginSubmit").click()
        time.sleep(10)
        driver.save_screenshot('d://JD//code.jpg')
        img = Image.open('d://JD//code.jpg')
        region = (660,212,901,320)
        nimg = img.crop(region)
        nimg.save("d://JD//code2.png")
        im = Image.open('d://JD//code2.png')
        width = im.size[0]  
        height = im.size[1]  
        pic=[]  
        for h in range(0, 108):  
            for w in range(0, 241):  
                pixel = im.getpixel((w, h))
                a=pixel[0]+pixel[1]+pixel[2]
                pic.append(a)
        pic2=[]
        for w in range(0, 241):  
            pixel = im.getpixel((w, 0))
            a=pixel[0]+pixel[1]+pixel[2]
            pic2.append(a)
        same=np.mean(pic2)
        for i in range(1,11):
            mid_1=str(i)+str(i)
            im2 = Image.open('d://JD//practice2//%s.png'%str(mid_1))
            pic2_1=[]
            for w in range(0, 241):  
                pixel = im2.getpixel((w, 0))
                a=pixel[0]+pixel[1]+pixel[2]
                pic2_1.append(a)
            if np.mean(pic2_1)==same:
                print(mid_1)
                break
        cus1=[]
        for h in range(0, 108):  
            for w in range(0, 241):  
                pixel = im.getpixel((w, h))
                a=pixel[0]+pixel[1]+pixel[2]
                cus1.append(a)
        ww=[]
        hh=[]
        i=0
        for h in range(0, 108):  
            for w in range(0, 241):  
                pixel = im2.getpixel((w, h))
                a2=pixel[0]+pixel[1]+pixel[2]
                if a2!= cus1[i]:
                    ww.append(w)
                    hh.append(h)
                i=i+1
        vbb_min=ww.index(min(ww))
        pixel = im.getpixel((ww[vbb_min]+16,hh[vbb_min]))
        aim_min=pixel[0]+pixel[1]+pixel[2]
        vbb_max=ww.index(max(ww))
        pixel = im.getpixel((ww[vbb_max]-16,hh[vbb_max]))
        aim_max=pixel[0]+pixel[1]+pixel[2]
        if aim_max<aim_min:
            aim_fin=ww[vbb_max]-20
        else:
            aim_fin=ww[vbb_min]+15
        aim_fin=aim_fin+23
        slideblock = driver.find_element_by_xpath('//xxx/div/div/div/div[2]/div[3]')
        ActionChains(driver).click_and_hold(slideblock).perform()   
        ActionChains(driver).move_by_offset(xoffset=aim_fin-1, yoffset=0).perform()
        time.sleep(0.4)
        ActionChains(driver).release(slideblock).perform()
        time.sleep(10)
        source=driver.page_source
        source1=re.search('我的店铺(.*?)class',source,re.S).group(1)
        print('成功登陆')
        chance=0
    except:
        chance=chance+1
        if chance>20:
            driver.quit()
            print('京东店铺登陆出错了')
        else:
            driver.quit()
            login(2)
```
